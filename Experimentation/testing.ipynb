```python
import cv2 as cv
import numpy as np
print(cv.__version__)

# import a picture 
img = cv.imread('C:\\Users\\johan\\computervision Learning\\ComputrvisionLearning\\Experimentation\\pictures\\52b89327e0c751cb5f3c48aaba3d3ca1.jpg')
#cv.imshow('Original Image', img)
#cv.waitKey(0)

print("image shape",img.shape)
print("pixel",img[0,0])

new_img = img
new_img[:,:,1] = 0
cv.imshow('new Image', new_img)
cv.waitKey(0)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray Image -win anti old way mode', gray)
cv.waitKey(0)

edges = cv.Canny(img,100,200)
cv.imshow('Canny Edges', edges)

kernel = np.ones((3,3),np.uint8)
thickedges = cv.dilate(edges,kernel, iterations=2)
cv.imshow('Thick Edges', thickedges)
cv.waitKey(0)
cv.destroyAllWindows()
```
